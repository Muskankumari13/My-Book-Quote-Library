```python
""" Make a class for Books that have Function like :
Add Book,
view Books,
delete book,
search book,
update book,
"""
import os
from collections import Counter

class Reading_journal:
  def __init__(self):
    # make a list for store book information
    self.books=[]

  def AddBook(self):
    title=input("Enter the tittle of the book: ")
    Author=input("enter the author name of the book :")
    for book in self.books:
      if book['title'].lower()==title.lower() and book['Author'].lower()==Author.lower():
        print("book is already exists")
        return
    year=int(input("enter the year in which book is published:"))
    genre=set(input("Enter Genre:").split(','))
    status=input("enter status from this (read, completed , on hold): ")
    book_entry={
        'title':title,
        'Author':Author,
        'year':year,
        'genre':genre,
        'status':status
    }
    self.books.append(book_entry)
    print("book added successfully!")

  def view_All_Book(self):
    if not self.books:
      print("no book is present in the journal")
    #It goes through all books sorted by their published year from oldest to newest.
    for book in sorted(self.books, key=lambda x: x['year']):
      print(f"Title:{book['title']}")
      print(f"Author:{book['Author']}")
      print(f"Year:{book['year']}")
      print(f"Genre:{book['genre']}")
      print(f"status:{book['status']}")

  def search_book(self, genre, status):
    for book in self.books:
      if genre in book['genre'] or status == book['status']:
        print(f"Title:{book['title']}")
        print(f"Author:{book['Author']}")
        print(f"Year:{book['year']}")
        print(f"Genre:{book['genre']}")
        print(f"status:{book['status']}")
        return
    print("book is not present")

  def update_book(self, genre, status):
    for book in self.books:
      if genre in book['genre'] and status == book['status']:
        book['genre']=set(input("Enter the updated Genre (comma separated): ").split(','))
        book['status']=input("Enter the updated status: ")
        print("book updated successfully")
        return
    print("book is not present")

  def delete_book(self, genre, status):
    for book in self.books:
      if genre in book['genre'] and status == book['status']:
        self.books.remove(book)
        print("book deleted successfully")
        return
    print("book is not present")

  """It opens "books.txt" to write, then for each book in the list,
   it converts the book to a string and writes it on a new line in the file.
  """
  def save_books_to_file(self):
        with open("books.txt", "w") as file:
            for book in self.books:
                file.write(str(book) + "\n")

  """It checks if "books.txt" exists, then opens it to read, and for each line in the file,
  it converts the line back to a book (using eval) and adds it to the book list.
  """
  def load_books_from_file(self):
    if os.path.exists("books.txt"):
      with open("books.txt", "r") as file:
        for line in file:
          self.books.append(eval(line.strip()))

  def Book_completed_in_given_year(self, year):
    found=False
    for book in self.books:
      if book['status'].lower()=="completed" and book['year']==year:
        print(f"Title: {book['title']}, Author:{book['Author']}")
        found=True
    if not found:
      print("no completed book in this year ")

  def book_with_most_quotes(self):
    if not hasattr(self, 'quotes') or not self.quotes:
      print("no quotes are alvailable!")
      return
    title_counts=Counter([quote['book_title'] for quote in self.quotes])
    most_common=title_counts.most_common(1)[0]
    print(f"The book with the most quotes is '{most_common[0]}' with {most_common[1]} quotes.")

  def author_with_most_books(self):
    if not self.books:
        print("No books available.")
        return
    author_counts = Counter([book['Author'] for book in self.books])
    max_books = max(author_counts.values())
    top_authors = [author for author, count in author_counts.items() if count == max_books]
    print("Author(s) with the most entries:")
    for author in top_authors:
        print(f"{author} ({author_counts[author]} books)")
```

```python
"""
make a class for quotes in book and add function like :
add quote,
delete quote,
show quote,
search quote
"""
class Quote_Collection:
  def __init__(self):
    self.quotes=[]

  def add_quote(self):
    text=input("enter the quote:")
    book_tittle=input("enter the book tittle:")
    page_number=int(input("enter the page number:"))
    self.quotes.append({
      'text': text,
      'book_title': book_tittle,
      'page_number': page_number
    })
    print("quote added successfully")


  def show_quotes(self):
    if not self.quotes:
      print("no Quote is present ")
      return
    for book in sorted(self.quotes, key=lambda x: x['page_number']):
      print(f"Quote:{book['text']}")
      print(f"Tittle:{book['book_title']}")
      print(f"page_number:{book['page_number']}")

  def search_quote(self, book_title):
    found = False
    for quote in self.quotes:
      if quote['book_title'].lower()==book_title.lower():
        print(f"Quote:{quote['text']}")
        print(f"Tittle:{quote['book_title']}")
        found = True
    if not found:
      print("quote is not present")

  def delete_quote(self, text , title):
    for quote in self.quotes:
      if quote['text'].lower()==text.lower() and quote['book_title'].lower()==title.lower():
        self.quotes.remove(quote)
        print("quote deleted successfully")
        return
    print("quote is not present")

  def save_quotes_to_file(self):
        with open("quotes.txt", "w") as file:
            for quote in self.quotes:
                file.write(str(quote) + "\n")

  def load_quotes_from_file(self):
        if os.path.exists("quotes.txt"):
            with open("quotes.txt", "r") as file:
                for line in file:
                    self.quotes.append(eval(line.strip()))
```

```python
def main():
  #We are creating objects of the classes Reading_journal and Quote_Collection.
  #Using these objects, we can now access the functions inside those classes like this:
    Reading_journal_instance = Reading_journal()
    quote_collection = Quote_Collection()
#This while loop will run until true inputs are given, means if input is from 1 to 9.
    while True:
        user_choice = input(
            "Enter a choice from the Library System:\n"
            "\nFrom Books:\n"
            "1. Add a new book\n"
            "2. View all books\n"
            "3. Search books by genre/status\n"
            "4. Update a book\n"
            "5. Delete a book\n"
            "\nFrom Quotes:\n"
            "6. Add a new quote\n"
            "7. View all quotes\n"
            "8. Search quotes by book title\n"
            "9. Delete a quote\n"
            "\nAnalysis Features\n"
            "10.List completed books in a year\n"
            "11.Book with the most quotes\n"
            "12.Author(s) with most entries\n"
            "0. Exit\n"
            "\nYour choice: "
        )

        if user_choice == '0':
            print("Exiting the program")
            break

        elif user_choice == '1':
            Reading_journal_instance.AddBook()
        elif user_choice == '2':
            Reading_journal_instance.view_All_Book()
        elif user_choice == '3':
            genre = input("Enter genre to search: ")
            status = input("Enter status to search: ")
            Reading_journal_instance.search_book(genre, status)
        elif user_choice == '4':
            genre = input("Enter genre of the book to update: ")
            status = input("Enter current status of the book: ")
            Reading_journal_instance.update_book(genre, status)
        elif user_choice == '5':
            genre = input("Enter genre of the book to delete: ")
            status = input("Enter status of the book to delete: ")
            Reading_journal_instance.delete_book(genre, status)
        elif user_choice == '6':
            quote_collection.add_quote()
        elif user_choice == '7':
            quote_collection.show_quotes()
        elif user_choice == '8':
            book_title = input("Enter book title to search quotes: ")
            quote_collection.search_quote(book_title)
        elif user_choice == '9':
            text = input("Enter quote text to delete: ")
            title = input("Enter book title of the quote: ")
            quote_collection.delete_quote(text, title)
        elif user_choice == '10':
          year = int(input("Enter year to list completed books: "))
          Reading_journal_instance.Book_completed_in_given_year(year)
        elif user_choice == '11':
          Reading_journal_instance.book_with_most_quotes()
        elif user_choice == '12':
          Reading_journal_instance.author_with_most_books()
        else:
            print("Invalid choice. Please enter a number from 0 to 9.")

main()
```
